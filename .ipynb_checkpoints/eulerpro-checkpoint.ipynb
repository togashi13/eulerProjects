```python
def sum_of_abundant(n,v):
    for i in range(1,n):
        if (v[i] > i and v[n-i] > n-i):
            return True
    return False
    
```

```python
#q27
import eulerlib, itertools
max = 0
aAns = 0
bAns = 0
for a in range(-999,1000):
    for b in range(-1000,1001):
        if get_consec_primes(a,b) > max:
            max = get_consec_primes(a,b)
            aAns = a
            bAns = b
            
```

```python
def get_consec_primes(a,b):
    for i in itertools.count():
        n = i*i + a*i + b
        if not eulerlib.prime_numbers.is_prime(n):
            return i
```

```python
def prev_permutation(arr):
	i = len(arr) - 1
	while i > 0 and arr[i - 1] <= arr[i]:
		i -= 1
	if i <= 0:
		return False
	j = len(arr) - 1
	while arr[j] >= arr[i - 1]:
		j -= 1
	arr[i - 1], arr[j] = arr[j], arr[i - 1]
	arr[i : ] = arr[len(arr) - 1 : i - 1 : -1]
	return True
```
